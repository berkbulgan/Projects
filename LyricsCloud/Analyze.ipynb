```python
import pandas as pd
import numpy as np
f = open("Files/Adele.txt", "r")
x = []
y = []
for line in f:
    for word in line.split():
        if word.lower() in x:
            ind = x.index(word.lower())
            y[ind] = y[ind]+1
        else:
            x.append(word.lower())
            y.append(1)
f.close()
a = np.array(x)
b = np.array(y)
df = pd.DataFrame({"Word" : a, "Count" : b})
df.to_csv("adele.csv", index=False)
```
